```python
import pandas as pd
import re
import nltk

from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
nltk.download('stopwords')
nltk.download('wordnet')
```

```python
data={"text":["I love NLP!It is amazing.","NLP is used in chatbots and  search engins.","I do not like boring lectures.","This NLP session is very interesting!"]}
df=pd.DataFrame(data)
df
```

```python
stop_words=set(stopwords.words("english"))
lemmatizer=WordNetLemmatizer()
def clean_text(text):
    text=text.lower()
    text=re.sub(r'[^a-z\s]','',text)
    words=text.split()
    words=[w for w in words if w not in stop_words]
    words=[lemmatizer.lemmatize(w)for w in words]
    return"".join(words)
```

```python
df["clean_text"]=df["text"].apply(clean_text)
df
```

```python
tfidf=TfidfVectorizer()
X=tfidf.fit_transform(df["clean_text"])
print("Vocabulary:\n",tfidf.get_feature_names_out())
print("\nTF-IDF Matrix:\n",X.toarray())
```

```python
tfidf_df=pd.DataFrame(X.toarray(),columns=tfidf.get_feature_names_out())
tfidf_df
```
